Optimización de funciones por el descenso de gradiente: Recuerde que el descenso del gradiente es un algoritmo de optimización que permite converger hacia el valor mínimo de una función mediante un proceso iterativo. En aprendizaje automático este método se utiliza para minimizar una función que mide el error de predicción del modelo en el conjunto de datos.
Para las siguientes funciones:

	 - $ f_{1}(x, y) = (x - 0,7)^{2}+(y - 0.5)^{2} con: x, y \in [-4 , 4] $
	 - $ f_{2}(x, y) = xe^{(-x^{2}-y^{2})}    con: x, y \in [-2 , 2] $

```python
def func1(x1, x2):
    return (x1 - 0.7)*(x1 - 0.7) + (x2 - 0.5)*(x2 - 0.5)

def func2(x1, x2):
    return x1 * math.pow(math.e, -(x1*x1) -(x2*x2))

def graficar_funcion_R3(funcion, titulo, inicio, fin):

    # Se definen las dimensiones del área a graficar
    x = y = np.arange(inicio, fin, 0.05)
    X, Y = np.meshgrid(x, y)
    zs = np.array([funcion(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
    Z = zs.reshape(X.shape)

    # Operaciones para el manejo del gráfico
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    surf = ax.plot_surface(X, Y, Z, cmap = cm.coolwarm)
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    plt.title(titulo)

    plt.show()

graficar_funcion_R3(func1, '(x1−0.7)^2 + (x2−0.5)^2', -4, 4)

def hessiano(x1, x2, A, B, C):
    return A(x1, x2)*C(x1, x2) - B(x1, x2)*B(x1, x2)

def evaluar(a, A, B, C):
    for elem in a:
        funct_A = A(elem[0], elem[1])
        funct_H = hessiano(elem[0], elem[1], A, B, C)

        if funct_A>0 and funct_H>0:
            print("Se tiene un mínimo local en (", elem[0], elem[1], ")")
        if funct_A<0 and funct_H>0:
            print("Se tiene un mínimo local en (", elem[0], elem[1], ")")
        if funct_H<0:
            print("Se tiene un punto de silla en (", elem[0], elem[1], ")")

def derivada_func1_x(x1, x2):
    return 2*(x1 - 0.7)

def derivada_func1_y(x1, x2):
    return 2*(x2 - 0.5)

def derivada_func1_xx(x1, x2):
    return 2

def derivada_func1_yy(x1, x2):
    return 2

def derivada_func1_xy(x1, x2):
    return 0

a = [[0.7, 0.5]]

evaluar(a, derivada_func1_xx, derivada_func1_yy, derivada_func1_xy)

#----

graficar_funcion_R3(func2, 'x1*e^(−x1^2−x2^2)', -2, 2)

def derivada_func2_x(x1, x2):
    return -(2*x1*x1 - 1)*math.pow(math.e, -(x1*x1) -(x2*x2))

def derivada_func2_y(x1, x2):
    return -2*x1*x2*math.pow(math.e, -(x1*x1) -(x2*x2))

def derivada_func2_xx(x1, x2):
    return (4*x1*x1*x1 - 6*x1) *math.pow(math.e, -(x1*x1) -(x2*x2))

def derivada_func2_yy(x1, x2):
    return 2*x1*(2*x2*x2 - 1) *math.pow(math.e, -(x1*x1) -(x2*x2))

def derivada_func2_xy(x1, x2):
    return 0
    (4*x2*x2 -2) *x2*math.pow(math.e, -(x1*x1) -(x2*x2))
    
def hessiano2(x1, x2):
    return derivada_func2_xx(x1, x2)*derivada_func2_xy(x1, x2) - derivada_func2_yy(x1, x2)*derivada_func2_yy(x1, x2)

    
a = [[math.sqrt(1/2), 0], [-math.sqrt(1/2), 0]]


evaluar(a, derivada_func2_xx, derivada_func2_yy, derivada_func2_xy)
```

2. (3 puntos) Implemente el algoritmo del descenso del gradiente con tensores de Pytorch

```python
def descenso_gradiente(num_epocas, coef_aprendizaje, init_a, init_b, init_c, x_train_tensor, y_train_tensor):
    
    a = init_a
    b = init_b
    c = init_c
    
    print('Entrenando con el coeficiente de aprendizaje:', coef_aprendizaje)
    flag = True
    a_aux = -99999
    b_aux = -99999
    c_aux = -99999
       
    # Convergencia 
    for epoch in range(num_epocas):

        # Salida de la función 
        zhat = func1(a, b) + c * (x_train_tensor+y_train_tensor)#a + b * x_train_tensor

        #Cálculo del error 
        error = z_train_tensor - zhat

        # El gradiente calculado de forma manual
        a_grad = -2 * error.mean()
        b_grad = -2 * (x_train_tensor * error).mean()
        c_grad = -2 * ((x_train_tensor+y_train_tensor) * error).mean()

        # Verificamos los gradientes.
        #print("Gradiente de a", a_grad)
        #print("Gradiente de b", b_grad)

        # Se actualiza los parámetros 
        # Se necesita la opción NO_GRAD para no afectar el grafo de computación dinámica 
        # que mantiene Pytorch.
        with torch.no_grad():
            a -= coef_aprendizaje * a_grad
            b -= coef_aprendizaje * b_grad
            c -= coef_aprendizaje * c_grad
            
            if flag and a_aux == a and b_aux == b and c_aux == c:
                flag = False
                print('Se llegó al punto de convergencia en la época', epoch)
            a_aux = a
            b_aux = b
            c_aux = b
            
            
    return a, b, c
```
